```python
song = []
artist = []
genre = []
year = []
rank = []
for i in range(51):
    chart_p = billboard.ChartData('alternative-songs', year=1970+i)
    for x in range(len(chart_p)):
        song.append(chart_p[x].title)
        artist.append(chart_p[x].artist)
        genre.append('Alternative')
        year.append(1970+i)
        rank.append(chart_p[x].rank)
alternative = {"Song": song, "Artist": artist, "Genre": genre, "Year": year, "Rank": rank}
```
